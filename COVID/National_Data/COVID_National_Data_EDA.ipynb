# EDA: COVID Projections for US States

This notebook performs EDA on a dataset of projections for the spread of COVID-19 within the US. The data is cleaned up to focus on the changing prevalence of the virus within each state over the next three months (May to August). A severity threshold is established whereby the situation in a state is considered to be severe if there is more than one new infection case per 10,000 residents within a single day.

> Dataset: IHME COVID-19 Projections (May 12 update) <br>
> Source: http://www.healthdata.org/covid/data-downloads <br>
> Units: The 50 US states <br>
> Dates: May 12 - August 5


```python
# import libraries

import numpy as np
import pandas as pd
import datetime

import matplotlib.pyplot as plt

# Import the us library (after installation)
# See documentation here: https://pypi.org/project/us/
import us
```

```python
# read in the IHME projection data

ihme = pd.read_csv("./Raw_Data/Projected_Data/IHME_projections_05_12.csv")
print(ihme.shape)
ihme.head()
```

```python
# Edit the rows

# Make a list of all US state names as strings
US_states = [str(s) for s in us.states.STATES]

# Drop rows for locations outside the 50 US states.
ihme = ihme[ihme['location_name'].isin(US_states)]

# Rename the location column as states
ihme = ihme.rename(columns={'location_name': 'state'})

# reset the index
ihme.reset_index(inplace=True)
ihme.drop(columns=['index'], inplace=True)

# Show the results
print(ihme.shape)
ihme.head()
```

```python
# Edit the columns.

# Make a list of the relevant columns, focusing on infections.
cols = ['state', 
        'date',
        'est_infections_mean']

# Drop all columns except the relevant ones.
ihme = ihme[cols]

# Rename the infections column
ihme = ihme.rename(columns={'est_infections_mean': 'est_new_infections'})

# Show the results
print(ihme.shape)
ihme.head()
```

```python
# Check for missing values

ihme.isna().sum()
```

```python
# Edit the date column

# Convert the date column from string to date-time dtype.
ihme["date"] = pd.to_datetime(ihme['date'])

# Select only the date range from 5/12 to 8/5
start_date = datetime.datetime(2020, 5, 12)
end_date = datetime.datetime(2020, 8, 5)
ihme = ihme.loc[(ihme['date'] >= start_date) & (ihme['date'] < end_date)]

# reset the index
ihme.reset_index(inplace=True)
ihme.drop(columns=['index'], inplace=True)

# Show the results
print(ihme.shape)
ihme.head()
```

```python
# Check again for missing values

ihme.isna().sum()
```

```python
# Visualize the distribution of infections

plt.hist(ihme["est_new_infections"]);
```

```python
# Which states are projected to have the most new daily infections?

ihme.groupby(['state'])['est_new_infections'].sum().sort_values(ascending=False)
```

```python
# Visualize the total number of new daily infections by state

plt.figure(figsize=(8,12))
ihme.groupby(['state'])['est_new_infections'].sum().sort_values().plot.barh();
```

```python
# Plot the change in daily new infections for some states

penn = ihme.loc[ihme['state'] == 'Pennsylvania']
california = ihme.loc[ihme['state'] == 'California']

penn.plot(x='date', y='est_new_infections', title="Pennsylvania")
california.plot(x='date', y='est_new_infections', title ="California");
```

```python
# Add state populations to the dataframe

# Read in state population data from the US Census Bureau
pops = pd.read_csv('./Raw_Data/US_state_population_data.csv')

# Select only the columns for state name and 2019 population size
pops = pops[['NAME', 'POPESTIMATE2019']]

# Rename the columns for state and population
pops = pops.rename(columns={'NAME': 'state', 'POPESTIMATE2019': 'est_population'})

# Drop rows for territories outside the 50 states
pops = pops.loc[pops['state'].isin(US_states)]

# Set the index to be the column for state names
pops.set_index('state', inplace=True)

# Add the population column to the IHME dataframe.
ihme = ihme.join(pops, on='state', how='left')

# Show the results
print(ihme.shape)
ihme.head()
```

```python
# Add a column for infection rate:
# number of infections per 10,000 residents

ihme["daily_infection_rate"] = (ihme["est_new_infections"] / ihme["est_population"]) * 10_000

# Show the results
print(ihme.shape)
ihme.head()
```

```python
# Visualize the distribution of daily infection rates

plt.hist(ihme['daily_infection_rate']);
```

```python
# Which states are projected to have 
# the highest average daily infection rates?

ihme.groupby(['state'])['daily_infection_rate'].mean().sort_values(ascending=False)
```

```python
# Visualize the states with the highest average daily infection rates
plt.figure(figsize=(8,12))
ihme.groupby(['state'])['daily_infection_rate'].mean().sort_values().plot.barh();
```

```python
# Create a severity indicator based on daily infection rate

# Set a threshold to define severe covid conditions
# as one new infection per 10,000 local residents.
thresh = 1

# Add a column to indicate severe covid conditions,
# in which 1 means severe, 0 means not severe.
ihme['severe'] = np.where(ihme['daily_infection_rate'] >= thresh, 1, 0)

# Display results
print(ihme['severe'].value_counts())
ihme.head()
```

```python
# Which states are projected to have the most total days
# with severe covid conditions?

ihme.groupby(['state'])['severe'].sum().sort_values(ascending=False)
```

```python
# Visualize the states with the most total days of severe covid conditions

plt.figure(figsize=(8,12))
ihme.groupby(['state'])['severe'].sum().sort_values().plot.barh()
plt.title("Total Days of Severe COVID");
```

```python
# Export the dataframe to csv file

ihme.to_csv('./Cleaned_Data/COVID_proj_US_512.csv', index=False)
```
