### This notebook evaluates the accuracy of the IHME projections for COVID daily new infections (from May 4th through May 11th)
> The focus is on the period from May 4th to May 11th because on May 4th IHME made a major change to its model, and our projections start on May 12th. So there is only this one week of data for which we have both projections and historical data for IHME's current model. <br><br>
> The results are not great. IHME predicted a much steeper and more steady decline in new daily infections across the country than actually occurred. This probably has to do with states opening up more quickly than expected during this time period.

```python
# import libraries

import numpy as np
import pandas as pd
import us
import datetime

from sklearn import metrics 

import matplotlib.pyplot as plt
```

### Get the predictions

```python
# read in the May 4th predictions

pred_df = pd.read_csv('./Cleaned_Data/COVID_proj_US_504.csv')

pred_df.head()
```

```python
# Select the rows and columns from the prediction data:
# Rows: dates from May 4th through May 11th
# Columns: focus on est_infections

# filter for the dates from May 4th through May 11th
pred_df = pred_df.loc[(pred_df['date'] >= '2020-05-04') & 
                      (pred_df['date'] <= '2020-05-11')]

pred_df = pred_df[['state', 'date', 'est_infections']]

print(pred_df.shape)
pred_df.head()
```

```python
# Set a variable for the predicted values

preds = pred_df['est_infections']
```

### Get the actual values

```python
# read in the May 13th dataset

ihme = pd.read_csv("./Raw_Data/Projected_Data/IHME_projections_05_12.csv")

print(ihme.shape)
ihme.head()
```

```python
# Edit the rows

# Make a list of all US state names as strings
US_states = [str(s) for s in us.states.STATES]

# Drop rows for locations outside the 50 US states.
ihme = ihme[ihme['location_name'].isin(US_states)]

# Rename the location column as states
ihme = ihme.rename(columns={'location_name': 'state'})

# reset the index
ihme.reset_index(inplace=True)
ihme.drop(columns=['index'], inplace=True)

# Show the results
print(ihme.shape)
ihme.head()
```

```python
# Edit the columns.

# Make a list of the relevant columns, focusing on infections.
cols = ['state', 
        'date',
        'est_infections_mean']

# Drop all columns except the relevant ones.
ihme = ihme[cols]

# Rename the infections column
ihme = ihme.rename(columns={'est_infections_mean': 'est_new_infections'})

# Show the results
print(ihme.shape)
ihme.head()
```

```python
# Edit the date column

# Convert the date column from string to date-time dtype.
ihme["date"] = pd.to_datetime(ihme['date'])

# Select only the date range from 5/4 to 5/11
start_date = datetime.datetime(2020, 5, 4)
end_date = datetime.datetime(2020, 5, 11)
ihme = ihme.loc[(ihme['date'] >= start_date) & (ihme['date'] <= end_date)]

# reset the index
ihme.reset_index(inplace=True)
ihme.drop(columns=['index'], inplace=True)

# Show the results
print(ihme.shape)

ihme.head()
```

```python
# set a variable for actual historical values

hist_df = ihme
actual = hist_df['est_new_infections']
```

### Evaluate the accuracy of the predictions

```python
# Calculate the RMSE

rmse = np.sqrt(metrics.mean_squared_error(preds, actual))
print('RMSE', rmse)

# Calculate the r2 score

r2 = metrics.r2_score(preds, actual)
print('R2 Score', r2)
```

```python
# Visualize the difference between projected and actual infections

fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Projected vs. Actual Total Estimated Infections Over Time')
ax1.plot(pred_df.groupby('date').sum())
ax2.plot(hist_df.groupby('date').sum())

ax1.set(ylabel='Projected')
ax2.set(ylabel='Actual')

fig.autofmt_xdate();
```

```python
# Calculate the daily error for total infections in the country

pred_sum = pred_df.groupby('date').sum()
hist_sum = hist_df.groupby('date').sum()
residuals = pred_sum['est_infections'] - hist_sum['est_new_infections']
residuals
```

```python
# Visualize the absolute daily error for total infections in the country

plt.plot(residuals);
```
