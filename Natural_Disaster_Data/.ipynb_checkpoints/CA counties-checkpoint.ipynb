```python
import pandas as pd
```

```python
df = pd.read_csv('Map_data.csv') 
df.head(2)
```

```python
df.isnull().sum()
```

```python
pd.DataFrame(df.groupby('County')['Incident Type'].value_counts())
```

```python
df['IncidentBeginDate'] = pd.to_datetime(df['Incident Begin Date'])
```

```python
df['BeginMonth'] = df['IncidentBeginDate'].dt.month
df['BeginYear'] = df['IncidentBeginDate'].dt.year
df['BeginDay'] = df['IncidentBeginDate'].dt.day
```

```python
disasters = df[(df['BeginYear'] >= 2010) & (df['Incident Type'] != 'Biological')][['Incident Type', 'County', 'BeginYear', 'BeginMonth']]
disasters.dropna(inplace = True)
```

```python
disasters
```

```python
organized = pd.DataFrame({'count' : disasters.groupby( [ "County", 'BeginYear', 'BeginMonth', "Incident Type"] ).size()}).reset_index()
organized
```

```python
total = pd.DataFrame(organized.groupby(['County', 'BeginYear', 'BeginMonth', 'Incident Type'])[['count']].sum().sort_values(by = ['BeginYear', 'BeginMonth']))
total = total.reset_index()
#count represent 10 years of data
total['countperyear'] = total['count'] / 10
total
```

```python
total[(total['Incident Type'] == 'Fire')]
```

```python
total.shape
```

```python
final = total[total['BeginMonth'].isin([5, 6, 7])]
final
```

```python
final.shape
```

```python
#per incident count per month taking into account 10 years
fi = pd.DataFrame({'count' : final.groupby( ['BeginMonth','County', "Incident Type"] ).size()}).reset_index()
fi.sort_values(['BeginMonth', 'County', 'count'], ascending = [True, True, False], inplace =True)
fi.head(60)
```

```python
fi[(fi['County'] == 'Mendocino County') & (fi['Incident Type'] == 'Fire')]
```

```python
#nationwide
national = organized.groupby(['BeginYear', 'BeginMonth', 'Incident Type'])[['count']].sum()
national.head(50)
```
