```python
# summary of all states (combined from map_data above)

dis_dec_sums = pd.read_csv('./../Natural_Disaster_Analysis/DisasterDeclarationsSummaries.csv')
print(dis_dec_sums.shape)
dis_dec_sums.head(3)
```

```python
dis_dec_sums.columns
```

```python
dis_dec_sums.isnull().sum().sort_values(ascending = False)
```

```python
pd.set_option('display.max_rows', 500)
pd.DataFrame(dis_dec_sums.groupby('state')['incidentType'].value_counts())
```

```python
dis_dec_sums['declarationDate'] = pd.to_datetime(dis_dec_sums['declarationDate'])
dis_dec_sums['incidentBeginDate'] = pd.to_datetime(dis_dec_sums['incidentBeginDate'])
dis_dec_sums['incidentEndDate'] = pd.to_datetime(dis_dec_sums['incidentEndDate'])
```

```python
dis_dec_sums['beginMonth'] = dis_dec_sums['incidentBeginDate'].dt.month
dis_dec_sums['beginYear'] = dis_dec_sums['incidentBeginDate'].dt.year
dis_dec_sums['beginDay'] = dis_dec_sums['incidentBeginDate'].dt.day
```

```python
dis_dec_sums['beginMonth'].value_counts()
```

```python
dis_dec_sums[dis_dec_sums['beginYear'] == 2020][['incidentBeginDate', 'disasterType', 'incidentType', 'state', 'declaredCountyArea']]
```

```python
disasters_2020 = dis_dec_sums[(dis_dec_sums['beginYear'] == 2020) & (dis_dec_sums['incidentType'] != 'Biological')][['disasterType', 'incidentType', 'declaredCountyArea', 'state', 'beginYear', 'beginMonth', 'beginDay']]
disasters_2020.dropna(inplace = True)

counties = []
for county in disasters_2020['declaredCountyArea']:
    if county.endswith('(County)'):
        counties.append(county[:-9])
    else:
        counties.append(county)
    
```

```python
disasters_2020['declaredCountyArea'] = counties
```

```python
disasters_2020
```

```python
disasters_2020.drop(index = 57357, inplace = True)
```

```python
!pip install geopy
```

```python
import time
from geopy.geocoders import Nominatim
```

```python
disasters_2020['address'] = disasters_2020['declaredCountyArea'] + ' County, ' + disasters_2020['state']
```

```python
disasters_2020.head(3)
```

```python
geolocator = Nominatim(user_agent="griffin_app_1")
address_latlong = []

for address in disasters_2020['address']:
    location = geolocator.geocode(address)
    address_latlong.append([address, location.latitude, location.longitude])
    time.sleep(1)
    
address_latlong[:5]
```

```python
address_latlong = pd.DataFrame(address_latlong)
```

```python
address_latlong.columns = ['address', 'latitude', 'longitude']
```

```python
disasters_2020_latlong = pd.merge(disasters_2020, address_latlong, on = 'address')
```

```python
disasters_2020_latlong.head(5)
```

```python
disasters_2020_latlong.state.value_counts()
```
